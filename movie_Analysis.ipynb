```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/moviesBasicData.csv")

# Display the first few rows
df.head()
```

```python
# View column names and data types
df.info()
```

```python
# Get summary statistics
df.describe(include='all')
```

```python
# Group by genre and calculate average rating
avg_rating = df.groupby('gen_title')['rev_rating'].mean().sort_values(ascending=False)

# Plot it
avg_rating.plot(kind='bar', figsize=(10,6), title='Average Rating by Genre')
plt.ylabel('Average Rating')
plt.xlabel('Genre')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Show column names
df.columns
```

```python
# Group by genre and calculate average rating
avg_rating = df.groupby('genre')['rating'].mean().sort_values(ascending=False)

# Plot it
avg_rating.plot(kind='bar', figsize=(10,6), title='Average Rating by Genre')
plt.ylabel('Average Rating')
plt.xlabel('Genre')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
movies_per_year = df['year'].value_counts().sort_index()

movies_per_year.plot(kind='line', figsize=(10,6), title='Movies Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.grid(True)
plt.show()
```

```python
avg_num_ratings = df.groupby('genre')['num_rating'].mean().sort_values(ascending=False)

avg_num_ratings.plot(kind='bar', figsize=(10,6), title='Average Number of Ratings by Genre')
plt.ylabel('Average Number of Ratings')
plt.xlabel('Genre')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
